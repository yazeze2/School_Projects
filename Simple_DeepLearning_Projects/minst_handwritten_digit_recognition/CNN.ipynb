```python
# Training/Validation and Testing Data
tr_image_file = '../data/train-images-idx3-ubyte'
tr_label_file = '../data/train-labels-idx1-ubyte'
ts_image_file = '../data/t10k-images-idx3-ubyte'
ts_label_file = '../data/t10k-labels-idx1-ubyte'

tr_images = idx2numpy.convert_from_file(tr_image_file)
tr_labels = idx2numpy.convert_from_file(tr_label_file)
ts_images = idx2numpy.convert_from_file(ts_image_file)
ts_labels = idx2numpy.convert_from_file(ts_label_file)

tr_images.shape, tr_labels.shape, ts_images.shape, ts_labels.shape
```

```python
#Checking Class Imbalance
class_counts = Counter(tr_labels).most_common()
x = [i[0] for i in class_counts]
y = [i[1] for i in class_counts]
plt.bar(x,y,color='b')
plt.title('Class Counts in Training Dataset');
```

```python
X_tr, X_val, y_tr, y_val = train_test_split(tr_images/255.0,tr_labels,test_size=0.1)

X_tr.shape, y_tr.shape, X_val.shape, y_val.shape
```

 - ### Tune Network/Parameters Manually using Loss Curves

```python
from tensorflow.keras.callbacks import TensorBoard
```

```python
#Use Tensorboard to monitor the values of at each layer - checking for Vanishing/Exploding gradients
tensorboard = TensorBoard(log_dir = 'tensor_logs/', histogram_freq=2)
```

```python
model = Sequential()
model.add(Conv2D(64, (3,3), input_shape = (28,28,1), kernel_initializer='he_uniform', activation='relu'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Conv2D(32, (3,3), kernel_initializer='he_uniform', activation='relu'))
model.add(MaxPooling2D((2,2),padding='same'))
model.add(Flatten())
model.add(Dense(784,kernel_initializer='he_uniform',activation='relu'))
model.add(Dropout(0.4))
model.add(Dense(10,activation='softmax'))

sgd = tf.keras.optimizers.experimental.SGD(
    learning_rate=0.075,
    momentum=0.05)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.summary()
```

```python
history = model.fit(X_tr, y_tr, validation_data=(X_val, y_val),
                  batch_size = 256, epochs=20,
                  callbacks=[early_stopping, tensorboard], verbose=True)
```

```python
fig, axs = plt.subplots(1,2, figsize=(16,6))

axs[0].plot(history.history['loss'], label='train', marker='*')
axs[0].plot(history.history['val_loss'], label='validation',marker='*')
axs[0].set_title('Loss')
# axs[0].set_ylim(0, 1)
axs[0].legend()

axs[1].plot(history.history['accuracy'], label='train',marker='*')
axs[1].plot(history.history['val_accuracy'], label='validation',marker='*')
axs[1].set_title('Accuracy')
axs[1].legend()

plt.tight_layout();
```

```python
training_accuracy = history.history['accuracy'][-1]
validation_accuracy = history.history['val_accuracy'][-1]

training_accuracy, validation_accuracy
```
